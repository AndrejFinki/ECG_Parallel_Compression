```python
import pandas
import matplotlib.pyplot
import numpy
```

```python
def display_results_standard( data: pandas.DataFrame, title: str ):
    for test in data['File']:
        matplotlib.pyplot.plot(
            data[ data.File == test ].columns[1:],
            data[ data.File == test ].iloc[0, 1:],
            marker = 'o',
            linestyle = 'dashed',
            label = test
        )
    
    matplotlib.pyplot.title( title )
    matplotlib.pyplot.xlabel( "MPI Processes" )
    matplotlib.pyplot.ylabel( "Execution time (ms)" )
    
    matplotlib.pyplot.legend(
        loc = 'center left',
        bbox_to_anchor = ( 1, 0.5 )
    )

    matplotlib.pyplot.tight_layout()
    matplotlib.pyplot.show()
```

```python
def display_results_log( data: pandas.DataFrame, title: str ):
    for test in data['File']:
        matplotlib.pyplot.plot(
            data[ data.File == test ].columns[1:],
            data[ data.File == test ].iloc[0, 1:],
            marker = 'o',
            linestyle = 'dashed',
            label = test
        )
    
    matplotlib.pyplot.title( title )
    matplotlib.pyplot.xlabel( "MPI Processes" )
    matplotlib.pyplot.ylabel( "Execution time (ms)" )
    matplotlib.pyplot.yscale( 'log' )
    
    matplotlib.pyplot.legend(
        loc = 'center left',
        bbox_to_anchor = ( 1, 0.5 )
    )

    matplotlib.pyplot.tight_layout()
    matplotlib.pyplot.show()
```

```python
def display_results_transposed( data_o: pandas.DataFrame, title: str ):
    data = data_o.transpose()
    data.columns = data.iloc[0]
    data = data[1:]
    for index, row in data.iterrows():
        matplotlib.pyplot.plot(
            [ col[:-2] for col in data.columns ],
            row,
            marker = 'o',
            linestyle = 'dashed',
            label = index
        )
    
    matplotlib.pyplot.title( title )
    matplotlib.pyplot.xlabel( "File Size (MB)" )
    matplotlib.pyplot.ylabel( "Execution time (ms)" )
    matplotlib.pyplot.yscale( 'log' )
    
    matplotlib.pyplot.legend(
        loc = 'center left',
        bbox_to_anchor = ( 1, 0.5 )
    )

    matplotlib.pyplot.tight_layout()
    matplotlib.pyplot.show()
```

```python
data = pandas.read_csv( '../results/2core.csv' )
display_results_standard( data, '2-Core Machine' )
display_results_log( data, '2-Core Machine Log' )
display_results_transposed( data, '2-Core Machine (Transposed)' )
```

```python
data = pandas.read_csv( '../results/4core.csv' )
display_results_standard( data, '4-Core Machine' )
display_results_log( data, '4-Core Machine Log' )
display_results_transposed( data, '4-Core Machine (Transposed)' )
```
