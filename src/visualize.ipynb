```python
import pandas
import matplotlib.pyplot
```

```python
data = pandas.read_csv( '../results/results.csv' )
```

```python
data
```

```python
data_100 = data.loc[ data.Test == '100.ecg' ].drop( columns = ['Test'] ).reset_index( drop = True )
data_119 = data.loc[ data.Test == '119.ecg' ].drop( columns = ['Test'] ).reset_index( drop = True )
```

```python
data_100_2core = data_100.loc[ data_100.Machine == '2-core' ].drop( columns = [ 'Machine' ] )
matplotlib.pyplot.title( "100.ecg on a 2-core machine" )
matplotlib.pyplot.xlabel( "MPI Process count" )
matplotlib.pyplot.ylabel( "Average algorithm run time (ms)" )
matplotlib.pyplot.plot(
    data_100_2core.columns,
    data_100_2core.iloc[0,:],
    marker = 'o',
    linestyle = 'dashed',
    color = 'orange',
)
```

```python
file_list = [ '100.ecg', '119.ecg' ]
machine_type_list = [ '2-core', '4-core', 'i5-12450H' ]
```

```python
file = file_list[0]
for machine_type in machine_type_list:
    current_data = data.loc[ data.Test == file ].drop( columns = ['Test'] ).reset_index( drop = True )
    current_data = current_data.loc[ current_data.Machine == machine_type ].drop( columns = [ 'Machine' ] )
    matplotlib.pyplot.title( f"{file} on a {machine_type} machine" )
    matplotlib.pyplot.xlabel( "MPI Process count" )
    matplotlib.pyplot.ylabel( "Average algorithm run time (ms)" )
    matplotlib.pyplot.plot(
        current_data.columns,
        current_data.iloc[0,:],
        marker = 'o',
        linestyle = 'dashed',
        label = machine_type
    )
matplotlib.pyplot.legend()
matplotlib.pyplot.show()
```
