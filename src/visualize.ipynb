```python
import pandas
import matplotlib.pyplot
import numpy
```

```python
data = pandas.read_csv( '../results/results.csv' )
data
```

```python
data_2_core    = data.loc[ data.Machine == '2-core' ].drop( columns = [ 'Machine' ] ).reset_index( drop = True )
data_4_core    = data.loc[ data.Machine == '4-core' ].drop( columns = [ 'Machine' ] ).reset_index( drop = True )
data_i5_12450H = data.loc[ data.Machine == 'i5-12g' ].drop( columns = [ 'Machine' ] ).reset_index( drop = True )
```

```python
def display_results_standard( data: pandas.DataFrame, title: str ):
    for test in data[ 'Test' ]:
        matplotlib.pyplot.title( title )
        matplotlib.pyplot.xlabel( "MPI Processes" )
        matplotlib.pyplot.ylabel( "Execution time (ms)" )
        matplotlib.pyplot.plot(
            data[ data.Test == test ].columns[ 1 : ],
            data[ data.Test == test ].iloc[ 0, 1 : ],
            marker = 'o',
            linestyle = 'dashed',
            label = test
        )
        matplotlib.pyplot.legend()
```

```python
def display_results_log( data: pandas.DataFrame, title: str ):
    for test in data[ 'Test' ]:
        matplotlib.pyplot.title( title )
        matplotlib.pyplot.xlabel( "MPI Processes" )
        matplotlib.pyplot.ylabel( "Execution time (ms)" )
        matplotlib.pyplot.yscale( 'log' )
        matplotlib.pyplot.plot(
            data[ data.Test == test ].columns[ 1 : ],
            data[ data.Test == test ].iloc[ 0, 1 : ],
            marker = 'o',
            linestyle = 'dashed',
            label = test
        )
```

```python
display_results_standard( data_2_core, '2-Core Machine' )
```

```python
display_results_log( data_2_core, '2-Core Machine ( w/ log )' )
```

```python
display_results_standard( data_4_core, '4-Core Machine' )
```

```python
display_results_log( data_4_core, '4-Core Machine ( w/ log )' )
```

```python
display_results_standard( data_i5_12450H, 'i5-12450H Machine' )
```

```python
display_results_log( data_i5_12450H, 'i5-12450H Machine ( w/ log ) ' )
```

```python
tr_data_2core = pandas.read_csv( '../results/results_transposed/2core.csv' )
tr_data_4core = pandas.read_csv( '../results/results_transposed/4core.csv' )
tr_data_i5    = pandas.read_csv(    '../results/results_transposed/i5.csv' )
```

```python
tr_data_i5
```

```python
def display_results_transposed( data: pandas.DataFrame, title: str ):
    for pcount in data[ 'Pcount' ]:
        matplotlib.pyplot.title( title )
        matplotlib.pyplot.xlabel( "Data Size (MB)" )
        matplotlib.pyplot.ylabel( "Execution time (ms)" )
        matplotlib.pyplot.plot(
            data[ data.Pcount == pcount ].columns[ 1 : ],
            data[ data.Pcount == pcount ].iloc[ 0, 1 : ],
            marker = 'o',
            linestyle = 'dashed',
            label = pcount
        )
        matplotlib.pyplot.legend()
```

```python
display_results_transposed( tr_data_2core, "2-Core Machine (Alternative Graph)" )
```

```python
display_results_transposed( tr_data_4core, "4-Core Machine (Alternative Graph)" )
```

```python
display_results_transposed( tr_data_i5, "i5-12450H Machine (Alternative Graph)" )
```
